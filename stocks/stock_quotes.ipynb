```python
import pandas as pd
from decimal import Decimal
from datetime import datetime
from thetadata_utils import curl
from thetadata import ThetaClient
from credentials import USERNAME, PASSWORD
client = ThetaClient(username=USERNAME, passwd=PASSWORD, timeout=10)
```

```python
config = {
    "stock_ticker": 'spy',
    "underlying_ticker": 'spx',
    "options_ticker": 'spxw',
    "start_date": datetime.datetime(2024, 1, 1),
    "end_date": datetime.datetime(2024, 1, 3),
    "ivl": 1000  * 60,
    "ivl_name": '1m'
}
my_date = datetime(2024, 6, 26)
```

#### Expirations

```python
path = '/data/td/options/expirations'
expirations = pd.read_csv(f"{path}/{config['options_ticker']}.csv.gz", index=False, compression='gzip')
expirations['date'] = pd.to_datetime(expirations['date'], format='%Y%m%d')
expirations
```

#### 1m/1s quotes

```python
for date in pd.date_range(start=config['start_date'], end=config['end_date']):
    date_str = date.strftime("%Y%m%d")

    if date.date() in expirations['date'].values:
        query = {
            "root": config['ticker'],
            "start_date": date_str,
            "end_date": date_str,
            "ivl": config['ivl'],
            "use_csv": "true"
        }
        
        ohlc = curl('hist/stock/ohlc', query)
        filename = f'/data/thetadata/{config['stock_ticker']}/{config['ivl_name']}/' + date_str + '.csv.gz'
        ohlc.to_csv(filename, compression='gzip')
        print(f"{date_str} saved to {filename}.")
```
